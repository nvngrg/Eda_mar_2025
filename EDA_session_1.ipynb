**Step-4**-->**provide index name**

```python
names=['ram','shyam','mohan','sohan'] # hard multipal column 
age=[20,19,28,25]
city=['Hyd','Blr','Sta','jbp']
pd.DataFrame(zip(names,age,city),index=['A','B','C','D'],columns=['Names','age','City'])
# N rows we need N columns name, same as for index
```

```python
# Q1) I want name and age as columns but city as index

names=['ram','shyam','mohan','sohan'] # hard multipal column 
age=[20,19,28,25]
city=['Hyd','Blr','Sta','jbp']
df=pd.DataFrame(zip(names,age),index=city,columns=['Names','age']) # remove city fdrom column and put to index
# zip only those columns that will show as cloumns 
# index = required column name(city here)
# coumns = only requiresd values (name and age)
```
