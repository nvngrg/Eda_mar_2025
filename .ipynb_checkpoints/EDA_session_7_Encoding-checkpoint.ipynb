## <center>**All at one encoding using map() encoding**</center>

- S1:- read dataset

- S2:- fatch cat columns using select_dtypes.columns() method

- use for loo to-

    - itrate unique cols_lable
 
    - itrate number for unique col_lables
 
    - make dictionary
 
    - replace columns value with numerical valuse
 
- S3:- Print encoded dataset 

```python
visa_df=pd.read_csv('Visadataset.csv')
cat_col=visa_df.select_dtypes(include='object').columns
for i in cat_col:
    uni_val=sorted(visa_df[i].unique())
    uni_lst=[i for i in range(len(uni_val))]
    uni_dict={key:value for key,value in zip(uni_val,uni_lst)}
    visa_df[i]=visa_df[i].map(uni_dict)
visa_df
```

```python
visa_df=pd.read_csv('Visadataset.csv')
```

```python
# step-1: Read the packege(pakege_name.class_name import method)
from sklearn.preprocessing import LabelEncoder
# step-2: save the packege/ call the mathod
le=LabelEncoder()
# step-3: apply fit transform on data
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
visa_df=pd.read_csv('Visadataset.csv')
cat_col=visa_df.select_dtypes(include='object').columns
for i in cat_col:
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df
```

# <center> **Fit and Transform**</center>

- fit means develop the logic

- transform means process the logic

- if we observe in map method first we developed dictionary logic

- the we appply that dictionary to column

- Developing the dictionary is kind of fit

- apply the dictionary to column means we are transforming the data cat to num

- whenever we are transforming or changing the data then use **fit transform**

- if we are not changing the data then only **fit**

## <center> **np.where** </center>

- using np.where also we can change the Categorical data to Numerical(also a encoding method)

- But **np.where is only for binary condition**

- np.where is **same as if-else**

- if condition is True : will apply True value

- if condition is False: will apply False value

- so **np.where use for only binary lables**

- **A column should have only two unique lables**

```python
# value=='certified' replace with 0
# otherwise 1 
visa_df=pd.read_csv('Visadataset.csv')
con=visa_df['case_status']=='Certified'
visa_df['case_status.csv']=np.where(con,0,1)
visa_df
```

# <center> **One hot encoder** </center>

- One hot means if **one will be ON , another will be OFF**

- `ON represents with 1`

- `Off represents with 0`

- For example case status has two unqiue lables

    - `Certified` 

    - `Denied`

- One hot encoder **creates new columns which is equal to number of uniuqe lables**

- For example `case status will create two new extra columns`

    - `case_status_Certified`

    - `case_status_Denied` 

|case_status|case_status_certified|case_status_denied|
|-|-|-|
|Denied|0|1|
|Certified|1|0|
|Denied|0|1|
