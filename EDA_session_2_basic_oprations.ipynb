**task-1**

- i want to create two list

- One is categorical column list

```python
visa_df.dtypes
```

**It is sowing result into a series**

- so, whenevere you want to create any list from series

- First convert the series into a dictionary

- now it becomes key:value pair

- itrate dictionary through for loop

- if value=='object' type, keep on string list

- else keep in numeric list

```python
data_type=visa_df.dtypes
dict(data_type)
```

```python
cat=[]
num=[]
for i,j in dict(data_type).items(): # The .items() method is used with dictionaries to return key-value pairs as tuples.
    if j=='object': # if value is object 
        cat.append(i) # add to cat list
    else:
        num.append(i) # otherwise add to num list
        
print(f"categorical columns are: {cat}")
print()
print(f"numerical columns are:{num} ")
```

### Using list comprehension

```python
string=[i for i,j in dict(data_type).items() if j=='object' ]
numerical=[i for i,j in dict(data_type).items() if j!='object' ]
print(f"categorical columns are: {string}")
print()
print(f"numerical columns are:{numerical} ")
```

### loc and iloc method

- usr to fatch dataset information using indexing

```python
# visa_df.iloc[<rows>,<columns>]
# visa_df.iloc[start:stop, start:stop]
# visa_df.iloc[start:stop:step, start:stop:step]
# visa_df.iloc[[list of rows],[list of columns]] 

# Case-1: rows=[100,200,300] columns=[2,3,7]
# Case-2: rows=[100:105] columnns=[2,5,7]
# case-3: rorws=[100:105] columns[2:7]
# case-4: rows=[100,200,300] columns=[2:7]
# case-5: rows=[100,200,300] column=:(all)
# case-6: rows=:(all) columns=[2,5,7]
```

```python
visa_df.iloc[[100,200,300],[2,5,7]]
# 100,200,300 row number and 2,5,7 column no.
```

```python
visa_df.iloc[100:105, [2, 5, 7]] 
# 100 to 105 rows with 2,5,7 clumns
```

```python
visa_df.iloc[100:105,2:7] 
# rows: 100 to 105 columns: 3 to 7
```

```python
visa_df.iloc[[100,200,300],2:7]
# row no: 100,200,300, columns: 2 to 7
```

```python
visa_df.iloc[[100,200,300],:]
# rows: 100,200,300, columns: all 
```

```python
visa_df.iloc[:,[2,5,7]]
# rows: all columns no: 2,5,7
```
